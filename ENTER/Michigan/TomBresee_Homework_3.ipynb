## Feature Analysis:

* We should be able to remove the 'datafile' column, as it references external .gz files which we don't care about.  However, we will keep this feature for a while as we use it to group by this feature, thus ensuring that data is all from the same individual exercise timespan.  
* We should be able to remove the three unknown data columns as well 

````python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
def kill_kol(your_col, your_df):
    if your_col in your_df.columns:
        your_df.drop(your_col,axis=1,inplace=True)
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
#kill_kol('datafile',df)
kill_kol('unknown_87',df)  #this column is literally only 0s and NaNs
kill_kol('unknown_88',df)  #this column is literally only 100,300, and NaNs
kill_kol('unknown_90',df)  #not needed for our analysis
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
````

* We also observe that the speed and enhanced speed are basically the same thing but by a factor of x1000, i.e. we should keep enhanced speed and remove 'speed' column:

````python
kill_kol('speed',df)  #this col is a repeat of enhanced speed (but by factor diff) 
````

* We observe that we have two different altitude features, but one can be deleted as the 'enhanced_altitude' is more accurate for our applications here.

````python
kill_kol('altitude',df)  #this col is a repeat of enhanced speed (but by factor diff) 
````

* Unfortunately, we do have a few features that are missing over HALF their values, so the right call is to delete those and focus on the features where we have a smaller fraction of missing features:
* Air Power, Form Power, Leg Spring Stiffness, Cadence, Ground Time, Power are missing significant counts of data, we will trim this out. 

````python
kill_kol('Air Power',df)
kill_kol('Cadence',df)
kill_kol('Power',df)
kill_kol('Form Power',df)
kill_kol('Ground Time',df)
kill_kol('Leg Spring Stiffness',df)
kill_kol('Vertical Oscillation',df)
````
